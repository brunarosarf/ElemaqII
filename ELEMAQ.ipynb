```python
#Verificação para Escoamento
Sigma_max_linha1 = sqrt(Sigma_a_linha1**2 + Tao_med_limha1**2)
print (f' A tensão máxima é {Sigma_max_linha1} MPa')

Fsy1 = (Sy / Sigma_max_linha1)
print (f' O fator de segurança em escoamento é {Fsy1}')
```

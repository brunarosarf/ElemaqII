# 1.0 Introdução


Engrenagens, eixos, chavetas e mancais são componentes essenciais na transmissão de potência e movimento em sistemas mecânicos. Esses elementos são amplamente utilizados em máquinas, veículos e equipamentos industriais, sendo fundamentais para garantir a eficiência e a confiabilidade dos mecanismos (JUVINALL; MARSHEK, 2013).

As engrenagens são dispositivos mecânicos dentados que possibilitam a transmissão de torque entre eixos, permitindo a modificação da velocidade e da direção do movimento (NORTON, 2013). Os eixos , por sua vez, desempenham a função de suportar cargas e transmitir potência rotativa aos demais componentes do sistema. Para garantir a fixação de elementos rotativos, como engrenagens e políticas, utilizam-se as chavetas , que evitam a penetração indesejada dos componentes acoplados (CHIAVERINI, 1986). Já os mancais são responsáveis ​​por sustentar eixos rotativos, redução de atrito e desgaste por meio de superfícies de superfície ou rolamento (JUVINALL; MARSHEK, 2013).

A seleção correta e a aplicação desses elementos são determinantes para o desempenho eficiente e seguro dos sistemas mecânicos, garantindo maior durabilidade e confiabilidade aos equipamentos (NORTON, 2013).

```python
#Dimensionamento a Fadiga

#Fator  de modificação de superfície Ka - Tabela 6-2 do Shigley
a2= 4.51
b2 = -0.265
Ka2 = a2 * Sut**b2
print (f' O fator Ka é {Ka2}')

#Fator modificador de tamanho Kb
Kb2 = 0.879*d2cor2**(-0.107)
print (f' O fator Kb é {Kb2}')

#Fator de modificação de carga Kc
Kc2 = 1
print (f' O fator Kc é {Kc2}')

#Fator de modificação de Temperatura Kd
Kd2 = 1
print (f' O fator Kd é {Kd2}')

#Fator de confiabilidade Ke
#Confiabilidade de 99%
Ke2 = 0.814
print (f' O fator Ke é {Ke2}')

#Fator de modificação por efeitos variados Kf
af2 = (139/Sut)**2
Kt2 = 1.62
Kf2 = 1 + (Kt2 - 1) / (1 + sqrt(af2/r2))
print (f' O fator Kf é {Kf2}')
Kts2 = 1.33
Kfs2 = 1 + (Kts2 - 1) / (1 + sqrt(af2/r2))
print (f' O fator Kfs é {Kfs2}')


#Limite de Endurance de Espécime Se'
Se_linha2 = 0.5 * Sut
print (f' O limite de endurance de espécime Se_linha é {Se_linha2} MPa')
Se2 = Ka2 * Kb2 * Kc2 * Ke2 * Kf2 * Se_linha2
print (f' O limite de endurance Se é {Se2} MPa')
```

```python
#Fator de Segurança em Fadiga pelo Critério de Goodman
sigma_a_flex2 = ((32 * Mmax2) / (pi * d2cor2**3)) * 1000
tao_med_tor2 = ((16 * T3) / (pi * d2cor2**3)) * 1000
print (f' A tensão de flexão é {sigma_a_flex2} MPa')
print (f' O Torque médio é {tao_med_tor2} Nm')

Sigma_a_linha2 = sigma_a_flex2 * Kf2
Tao_med_limha2 = sqrt(3*(tao_med_tor2 * Kfs2)**2)
print (f' Sigma linha de Flexão é {Sigma_a_linha2} MPa')
print (f' Torque linha médio é {Tao_med_limha2} Nm')


Fsf2 = 1 / ((Sigma_a_linha2/Se2) + (Tao_med_limha2/Sut))
print (f' O fator de segurança em fadiga é {Fsf2}')
```

```python
#Verificação para Escoamento
Sigma_max_linha2 = sqrt(Sigma_a_linha2**2 + Tao_med_limha2**2)
print (f' A tensão máxima é {Sigma_max_linha2} MPa')

Fsy2 = (Sy / Sigma_max_linha2)
print (f' O fator de segurança em escoamento é {Fsy2}')
```
